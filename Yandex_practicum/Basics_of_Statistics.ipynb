**ЦПТ** объясняет, почему нормальное распределение часто возникает в статистике, несмотря на распределение исходных данных. Если взять много независимых и однородных случайных величин, и посчитать их среднее, то независимо от формы распределения исходных данных распределение средних будет близким к нормальному.

**Центральная предельная теорема утверждает**, что:
- Если мы многократно извлекаем выборки размера n и вычисляем среднее значение для каждой выборки (aj), то распределение этих средних (A) значений будет приближаться к нормальному распределению;
- Среднее значение этого распределения будет практически совпадать со средним значением генеральной совокупности (μ);
- А стандартное отклонение распределения выборочных средних (будет меньше стандартного отклонения генеральной совокупности)

**Нулевая гипотеза**, обозначаемая как H0, это исходное предположение о данных.
**Альтернативная гипотеза**, обозначаемая как H1, представляет собой предположение, которое противоположно нулевой гипотезе.

**Z-тестом**, в котором для расчёта z-статистики использовали стандартную ошибку среднего (SE):

z = (x(среднее значение по выборке) − μ (предполагаемое среднее значение генеральной совокупности)) / SE, где SE = σ / корень из (n)

**k-статистика** = (среднее значение по выборке - предполагаемое среднее значение генеральной совокупности) / (стандартное отклонение генеральной совокупности / корень из (размер выборки))

Значит, **мы предполагали, что дисперсия генеральной совокупности известна**. Это и есть требование к данным для Z-теста. Мы можем рассчитать статистику и применить его только в случае, если известна дисперсия генеральной совокупности. 

Если мы выбираем стандартный уровень значимости α=0.05, то критические значения для z-статистики будут − 1.96 и 1.96.

**Ошибка первого рода** — это ситуация, когда исследователь отвергает нулевую гипотезу, хотя она на самом деле верна.

**Уровень значимости α** — это вероятность совершения ошибки первого рода, то есть отклонить нулевую гипотезу, когда она верна.

**P-value** — вероятность получить определённое или ещё более экстремальное значение статистического критерия при условии, что нулевая гипотеза верна. (или p-value — это вероятность того, насколько вероятно обнаружить текущие или более экстремальные значения статистики, если нулевая гипотеза верна)

Обновим порядок действий для **сценария использования p-value в рамках Z-теста:**
- формулируем H0, H1;
- выбираем α, находим соответствующие критические значения;
- считаем выборочное среднее;
- считаем z-статистику;
- считаем p-value;
- сравниваем p-value с уровнем значимости:
    - Если p-value меньше уровня значимости, отклоняем нулевую гипотезу.
    - Если p-value больше уровня значимости, то оснований отклонять нулевую гипотезу нет.

**Ошибка второго рода, β** — это ситуация, когда исследователь не отвергает нулевую гипотезу, хотя на самом деле верна альтернативная.

**Мощность статистического теста, 1−β** — это вероятность правильно отклонить нулевую гипотезу, когда она действительно неверна.

Примеры **ошибки первого рода:**
- Ваша нулевая гипотеза состоит в том, что вода чистая. Если ваш тест ошибочно обнаруживает загрязнение в чистой воде.
- Метод контроля качества ошибочно обнаруживает отклонение от стандарта у таблетки, которая на самом деле соответствует стандарту
- Вы сообщаете акционерам, что финансовый отчёт компании неправильный, хотя на самом деле он верный
- Ваша лаборатория разрабатывает новый тест на диагностику заболевания. Если тест показывает положительный результат у здорового человека

Примеры **ошибки второго рода:**
- Ваш тест не обнаруживает загрязнение в реально загрязнённой воде
- Ваш метод контроля качества не обнаруживает отклонение от стандарта у таблетки, которая на самом деле не соответствует стандарту
- Вы утверждаете, что финансовый отчёт компании правильный, хотя он на самом деле неправильный
- Если тест показывает отрицательный результат у больного человека

Примеры **мощности статистического теста:**
- Вероятность обнаружить загрязнение, когда оно действительно есть
- Вероятность обнаружить отклонение от стандарта, если оно действительно есть.
- Это позволит вам обнаружить неправильные финансовые отчёты, если они действительно такие
- Как называется вероятность того, что ваш тест правильно определит наличие заболевания у больного человека

Итак, **когда дисперсия неизвестна**, используют оценку дисперсии, основанную на выборке и вместо стандартной ошибки среднего SE получают оценку стандартной ошибки **ESE**. 

Изменение подхода к определению стандартной ошибки приводит к изменениям и в статистическом тесте: меняется формула статистического критерия и сам критерий. Статистику критерия теперь рассчитывают так: 

t = (x(среднее значение по выборке) − μ (предполагаемое среднее значение генеральной совокупности)) / ESE, где ESE = sx / корень из (n)

Для z-статистики получалось нормальное распределение, а для t-статистики получится **распределение Стьюдента**. Статистический тест, основанный на этом распределении, называется **T-тест**.
