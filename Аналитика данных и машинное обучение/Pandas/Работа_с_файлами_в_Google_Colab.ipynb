```python
# снова скачаем столбец Sex из датасета train в формате датафрейма
previous = pd.read_csv('/content/titanic.csv')[['Sex']]
previous.head() # по умолчанию первеы 5 строк
```

```python
# закодируем переменную через 0 и 1
pd.get_dummies(previous['Sex'], dtype = int).head(3)
```

```python
# удалим первый столбец, он избыточен
sex = pd.get_dummies(df['Sex'], drop_first = True)
sex.head(3)
```

```python
# сделаем то же самое для переменных Pclass и Embarked
embarked = pd.get_dummies(df['Embarked'], drop_first = True)
pclass = pd.get_dummies(df['Pclass'], drop_first = True)
```

```python
# присоединим закодированные через one-hot encoding переменные
# к исходному датафрейму через функцию .concat()
df = pd.concat([df, pclass, sex, embarked], axis = 1)
df.head(3)
```

Отбор признаков

```python
# удалим те столбцы, которые нам теперь не нужны
df.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Ticket', 'Embarked'], axis = 1, inplace = True)
df.head(3)
```

Нормализация данных

```python
# импортируем класс StandardScaler
from sklearn.preprocessing import StandardScaler

# создадим объект этого класса
scaler = StandardScaler()

# выберем те столбцы, которые мы хотим масштабировать
cols_to_scale = ['Age', 'Fare']

# рассчитаем среднее арифметическое и СКО для масштабирования данных
scaler.fit(df[cols_to_scale])

# применим их
df[cols_to_scale] = scaler.transform(df[cols_to_scale])

# посмотрим на результат
df.head(3)
```

```python
# некоторые названия столбцов теперь представляют собой числа, так быть не должно
df.columns
```

```python
# преобразуем эти переменные в тип str через функцию map()
df.columns = df.columns.map(str)
df.columns
```

3)Вывести статистику по каждому столбцу:

a)для категориальных переменных: количество уникальных значений, моду,
количество пропущенных значений


include=['object'] - выделить категориальную переменную

```python
category_cols = df.select_dtypes(include=['object']) #выбрали столбцы с категориальными переменными
category_cols_names = category_cols.columns.tolist()
print(f'категориальные столбцы: {category_cols_names}\n')
for col in category_cols_names:
    print(f'информация о столбце {col}:')
    print(f'количество уникальных значений: {category_cols[col].nunique()}') #для кол-ва уникальных волспользуемся функцией nunique()
    print(f'мода: {category_cols[col].mode().tolist()}') #для поиска моды - функцией mode() (Если в столбце несколько значений встречаются одинаковое количество раз, метод mode() вернет все такие значения.)
    print(f'количество пропущенных значений: {category_cols[col].isna().sum()}\n') #isna() - создания булевой маски, которая указывает, какие ячейки в столбце содержат пропущенные значения, sum() используется для подсчета количества True значений в маске
```

```python
df.describe(include='object')
```

b)Для интервальных: медиану, среднее, квантиль 90, квантиль 10, максимальное значение, минимальное значение, количество пропущенных значений

```python
category_int = df.select_dtypes(include=['int64','float64']) #выбрали столбцы с интервальные переменными
category_int_names = category_int.columns.tolist()
print(f'интервальные столбцы: {category_int_names}\n')

for col in category_int_names:
    print(f'информация о столбце {col}:')
    print(f'медиана: {category_int[col].median()}')
    print(f'среднее: {category_int[col].mean()}')
    print(f'квантиль 90: {category_int[col].quantile(0.9)}')
    print(f'квантиль 10: {category_int[col].quantile(0.1)}')
    print(f'максимум: {category_int[col].max()}')
    print(f'минимум: {category_int[col].min()}')
    print(f'кол-во пропусков: {category_int[col].isna().sum()}\n')
```

```python
df.describe(percentiles=[0.1, 0.5, 0.9], include=['int64','float64'])
```
