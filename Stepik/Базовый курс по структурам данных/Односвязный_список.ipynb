**Односвязный список**  —  представляет собой коллекцию элементов, каждый их которых содержит в себе хранимые данные, а также ссылку на связанный с ним элемент. Основным преимуществом данной структуры данных перед обычным массивом является **ее динамичность** — возможность легко менять количество элементов без расходования лишней памяти сверх той что требуется. То есть связный список **позволяет хранить произвольное число элементов в любой момент времени**. Как следствие, при инициализации связного списка нет необходимости объявлять его размер.

**Элементом односвязного списка** является **узел (Node)**. Он представляет из себя собственно сами данные, которые мы хотим хранить в нашей коллекции (данные могут быть любыми: число, строка, массив) и указатель (ссылку) на конкретный байт в памяти, где начинается другой узел.
