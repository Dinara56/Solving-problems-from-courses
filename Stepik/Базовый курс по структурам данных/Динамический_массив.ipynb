**Теория:**

**Динамический массив** — обертка над статическим массивом, содержащий методы для манипуляции с данными. Размер массива изменяется при добавлении нового элемента.

**Верные утверждения про динамический массив:**
- Получение элемента на произвольном месте выполняется за константное время, то есть O(1).
- Нахождение элемента по его значению выполняется за линейное время, то есть O(n).
- Добавление элемента  в любое место, хоть в конец выполняется  за линейное время, то есть O(n).
- Среднее (амортизированное) время добавления элемента  в конец выполняется равно O(1). Но в худшем случае выполняется за O(n).
- Удаление элемента в произвольном месте (не в конце) выполняется за линейное время, то есть O(n).
- Удаление элемента в конце  выполняется за константное время, то есть O(1).
