**Теория:**

**Хеш-функция** — функция (алгоритм), осуществляющая преобразование входных данных произвольной длины (обычно достаточно большой) в выходную строку установленной длины (достаточно маленькой). При этом достаточно уникальную строку, которая позволяет отличить выбранные входные данные от каких-то других данных. Преобразование, производимое хеш-функцией, называется хешированием.

Исходные данные называются входным массивом, ключом или сообщением. Результат преобразования (выходные данные) называется хешем или хеш-кодом.

Случай, при котором хеш-функция для разных ключей возвращает одинаковые хеш-коды, называется **коллизией.**

Существуют два основных способа реализации хеш-таблиц. Они различаются по способу разрешения коллизий: 

- Метод открытой адресации (закрытое хеширование). Стратегия разрешения коллизий: если ячейка занята, то переходим в следующую до тех пор, пока не найдем свободную ячейку, куда вставить элемент. 
 
- Метод цепочек (открытое хеширование). Стратегия разрешения коллизий: если ячейка занята, то все элементы с одинаковым хеш-кодом объединяются в список.
